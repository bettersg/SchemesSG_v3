```python
import pandas as pd
import numpy as np
import re
import string
```

```python
df_schemes = pd.read_csv('../df.csv', encoding='mac_roman')
df_schemes
```

```python
formatted = pd.DataFrame(columns=('Scheme','Description','Agency','Image','Link',"Who It's For", "What It Gives", 'Scheme Type'))

for i in range(len(df_schemes)): 
    for j in range(len(df_schemes.iloc[i]["Who's it for"].split(','))):
        for k in range(len(df_schemes.iloc[i]["What it gives"].split(','))):
            for l in range(len(df_schemes.iloc[i]["Scheme Type"].split(','))):
                new_row = {'Scheme':df_schemes.iloc[i]['Scheme'],
                               'Description':df_schemes.iloc[i]['Description'],
                               'Agency':df_schemes.iloc[i]['Agency'],
                               'Image':df_schemes.iloc[i]['Image'],
                               'Link':df_schemes.iloc[i]['Link'],
                               "Who It's For":df_schemes.iloc[i]["Who's it for"].split(',')[j],
                               "What It Gives":df_schemes.iloc[i]["What it gives"].split(',')[k],
                               "Scheme Type":df_schemes.iloc[i]['Scheme Type'].split(',')[l]}
                formatted = formatted.append(new_row, ignore_index=True)
```

```python
formatted.to_csv('formatted.csv')
```
